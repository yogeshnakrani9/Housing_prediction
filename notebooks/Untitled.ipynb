```python
import pandas as pd
```

```python
df = pd.read_csv("../input/train_cat.csv", index_col = "id")
```

```python
df.head()
```

```python
binary_cols = [c for c in df.columns if c.startswith("bin_")]
nominal_cols = [c for c in df.columns if c.startswith("nom_")]
ordinal_cols = [c for c in df.columns if c.startswith("ord_")]
cyclical_cols = ["day", "month"]
```

```python
for col in binary_cols:
    print(df[col].value_counts())
```

```python
for col in nominal_cols:
    print(df[col].value_counts())
```

```python
col_total = 0
for col in ordinal_cols:
    col_total += df[col].nunique()
print(col_total)
```

```python
for col in cyclical_cols:
    print(df[col].value_counts())
```
